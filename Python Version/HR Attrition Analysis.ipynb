```python
sns.countplot(x='Attrition', data = data)
plt.title("Over attrition rate")
plt.show()
```

```python
# Department Level Analysis
dept_attrition = data.groupby('Department')['Attrition'].value_counts(normalize = True).unstack()*100
dept_attrition
```

```python
sns.barplot(x= 'Department', y= 'Attrition', data=data)
plt.ylabel("Attrition %")
plt.xlabel("Attrition by Department")
plt.show()
```

```python
data.groupby('Attrition')['Age'].mean()
```

```python
# Gender Attrition 
gender_attrition = data.groupby('Gender')['Attrition'].value_counts(normalize = True).unstack()*100
gender_attrition
```

```python
# Job Satisfaction vs Attrition 
sns.countplot(x ='JobSatisfaction', hue = 'Gender', data = data)
plt.title("Attrition by Job Satisfaction")
plt.show()
```

# Insights
1. Attrition Rate = 16.1% , Sales has the highest Attrition
2. Employees who left are younger(avg age 29) vs those who stayed (avg age 37)
3. Males have slightly higher attrition than Females
4. Employees with low job satisfaction are 2x more likely to leave
